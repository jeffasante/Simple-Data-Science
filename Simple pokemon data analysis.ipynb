How powerful are a Fire-type Pokemon?

In the Pokemon game mechanics, Combat Power(CP) measure the strength of a Pokemon in a battle against another Pokemon.

A pokemon level can change by increase powering up during play mode. They have fixed values (Stamina, defense and attack) where it value cannot be changed.


```python
# code to how powerful are a Fire-type Pokemon on an average

pokemon_df[(pokemon_df.type1 == 'Fire') | (pokemon_df.type2 == 'Fire')].max_cp.mean()
```

```python
# code to how powerful are a Dark-type Pokemon on an average

pokemon_df[(pokemon_df.type1 == 'Dark') | (pokemon_df.type2 == 'Dark')].max_cp.mean()
```

it seems Dark-type pokemon has a 62% chance beat on Fire-type pokemon which has a chance of 38%.

## Plotting

Another	quick	way	to	get	a	feel	of	the	type	of	data	you	are	dealing	with	is	to	plot	a	histogram	for	each numerical	attribute.

```python
pokemon_df.info()
```

Since we cannot plot a boolean data, we have to use pandas drop object to remove the 'legendary' column. And also prevent matplotlib to display plot. 

```python
df = pokemon_df.drop('legendary',axis=1)
```

```python
df.head()
```

```python
df.hist(bins=50, figsize=(20,15));
```

let isolate the legendary column to a new DataFrame. 

```python
df_legendary_pokemon = pokemon_df[pokemon_df.legendary]
df_legendary_pokemon.head()
```

lets group them together by generation, calculate average max_cp for each group generation.

```python
avg_max_cp_by_generation = df_legendary_pokemon.groupby('generation').max_cp.mean()
avg_max_cp_by_generation
```

```python
# now, plot the series
avg_max_cp_by_generation.plot.bar(figsize=(20,15));
```

we can see that the 4 values are almost up close together, so lets change the y-axis limits to make the differences more pronounced.

```python
avg_max_cp_by_generation.plot.bar(ylim=(3200, 3500), figsize=(20,15));
```

on the average, the legendary Poemon in Generation 4 are the strongest while those in Generation 2 are the weakest. With Generation 1 and 2, theres a possibility that Generation 3 can defeat Generation 1.
